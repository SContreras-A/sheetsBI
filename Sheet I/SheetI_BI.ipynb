```python
import numpy

n = 100

x = numpy.ones(n)
print("Vx con método ones:", x)

y = numpy.arange(3,3*(n+1),3)
print("Vy con método arange: ", y)
```

```python
yT = 15 * numpy.sin(y)

y2 = numpy.round(yT).astype(int)
print("Indices de las entradas mayores a 6: ", numpy.where(y2>6))

print("¿Todas las entradas son mayores a 2? ", numpy.all(y2>2))

print("Cantidad de entradas iguales a 5: ",numpy.sum(y2==5))
```

##

```python
import pandas as pd

t1 = {'Name': ['Alexander', 'Jane'],
         'Age': [19, 36],
         'Sex': ['male', 'female'],
         'Profession': ['student', 'physician']}
df1 = pd.DataFrame(t1)

t2 = {'Name': ['Eric', 'Laura'],
         'Age': [22, 48],
         'Sex': ['male', 'female'],
         'Profession': ['lawyer', 'teacher']}
df2 = pd.DataFrame(t2)

t3 = {'Name': ['Peter', 'Julia'],
         'Age': [31, 24],
         'Sex': ['male', 'female'],
         'Profession': ['engineer', 'consultant']}
df3 = pd.DataFrame(t3)
```

## 

```python
dfUnido = pd.concat([df1, df2, df3], ignore_index=True)

print("Union DataFrame:")
print(dfUnido)
```

## 

```python
t4 = {'Name': ['Alexander', 'Jane', 'Eric', 'Laura', 'Peter', 'Julia'],
        'Size': [181, 162, 178, 183, 173, 169],
        'Salary': [600, 4200, 4000, 3400, 5200, 3600]}
df4 = pd.DataFrame(t4)

dfTotal = pd.merge(dfUnido, df4, on='Name', how='left')
print(dfTotal)
```

## 

```python
dfTotal['Size'] = dfTotal['Size'] - 1
print(dfTotal)
```

## 

```python
filtroEdad1 = dfTotal[dfTotal['Age'] > 30]
print("Filtro booleano: \n",filtroEdad1)

filtroEdad2 = dfTotal.query('Age > 30')
print("Filtro condicionado: \n",filtroEdad2)
```

## 

```python
dfOrdenadoSalario = dfTotal.sort_values(by='Salary', ascending=True)
print(dfOrdenadoSalario)
```

## 

```python
grupoPorSexo = dfTotal.groupby('Sex')

for sex, group_df in grupoPorSexo:
    print(f"Sex: {sex}\n{group_df}\n")
```

## 

```python
import numpy as np

data = csv[['sepal_length', 'sepal_width']]

def ecdf(x):
    return np.arange(1, len(x) + 1) / len(x)

sorted_length = np.sort(data['sepal_length'])
sorted_width = np.sort(data['sepal_width'])

length_grid, width_grid = np.meshgrid(sorted_length, sorted_width)
ecdf_grid = np.zeros_like(length_grid)

for i in range(length_grid.shape[0]):
    for j in range(length_grid.shape[1]):
        ecdf_val = np.mean((data['sepal_length'] <= length_grid[i, j]) & 
                           (data['sepal_width'] <= width_grid[i, j]))
        ecdf_grid[i, j] = ecdf_val

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(length_grid, width_grid, ecdf_grid, cmap='viridis')

ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('ECDF')

plt.show()
```
