```python
import matplotlib.pyplot as plt
import itertools
import pandas as pd
import numpy as np

datos = pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
D = (datos[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].values)

def drawPrincipalDirections(D):

    columnas = list(itertools.combinations(range(D.shape[1]), 2))

    for combinacion in columnas:
        
        Xi, Xj = combinacion
        

        dataX = D[:, [Xi, Xj]]
        
        m = list(sum(row)/len(row) for row in zip(*dataX))
        n = dataX.shape[0]
        cov_m = dataX - m
        cov = np.dot(cov_m.T, cov_m) / (n)
        
        eigenvalues, eigenvectors = np.linalg.eig(cov)


        plt.scatter(dataX[:, 0], dataX[:, 1])

        for i in range(2):
            plt.axline((m[0], m[1]), slope=eigenvectors[1, i] / eigenvectors[0, i] )
            
        plt.xlabel(f'Xi')
        plt.ylabel(f'Xj')
        plt.show()
drawPrincipalDirections(D)
```

CONCLUSION:

Una vez tenemos la gr√°fica, podemos notar la creaci√≥n de m√∫ltiples grupos de datos separados en m√∫ltiples combinaciones diferentes. Estos grupos significan que hay tendencias sobre algunas caracter√≠sticas. Las l√≠neas dibujadas nos permiten notar las medias y sus pendientes en los ejes X y Y, lo que permite identificar patrones de comportamiento de los datos y llegar a predecir futuros movimentos de los mismos.
